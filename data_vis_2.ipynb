```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
```

```python
# Sample data creation (replace this with your actual DataFrame)
df = pd.read_csv("weather.csv")
# df.drop('volume', axis=1, inplace=True)

```

```python
# Plotting the time series data
plt.figure(figsize=(10, 6))
ax1 = plt.gca()

# Plot the data for the first y-axis
for column in df.columns[1:3]:  # Assuming first two columns share the first y-axis
    ax1.plot(df.index, df[column], label=column, linewidth=0.5)

ax1.set_xlabel('Date')
ax1.set_ylabel('Values (Y1)')
ax1.grid(True)

# Create a second y-axis
ax2 = ax1.twinx()
for column in df.columns[5:7]:  # Assuming the third column is for the second y-axis
    ax2.plot(df.index, df[column], label=column, linewidth=0.5)

ax2.set_ylabel('Values (Y2)')

# Combine legends
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='best')

plt.title('Multiple Time Series Data with Dual Y-Axis')
plt.show()
```

```python
scaler = StandardScaler()


dt = df['date']
df_melted = df.drop(['date'],axis=1).iloc[:, :5]



df_melted = pd.DataFrame(scaler.fit_transform(df_melted), columns=df_melted.columns)



df_melted = df_melted.melt(value_vars=list(df_melted.columns), 
                    var_name='Type', value_name='Value')



plt.figure(figsize=(10, 6))
sns.violinplot(x='Type', y='Value', data=df_melted)

# Customizing the plot
plt.xlabel('Category')
plt.ylabel('Z-score Value')
plt.title('Data Distribution of Weather Data')


# Show plot
plt.show()
```
