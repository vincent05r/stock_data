```python
import yfinance as yf
import pandas as pd
from df_preprocessing import convert_to_pct
import os

import time
```

```python
target = 'GOOG'
target_ticker = yf.Ticker(target)
target_df = target_ticker.history(start='2010-01-01', end='2020-12-31', interval='1d')
```

```python
#processing 


target_df = target_df.reset_index()
```

```python
drop_list = ['Dividends', 'Stock Splits']
target_df.drop(drop_list, axis=1, inplace=True)
target_df = target_df.rename(columns= lambda x :  x.lower() )

#trim down to 1d inerval time
target_dt = pd.to_datetime(target_df['date'])
target_df['date'] = target_dt.dt.strftime('%Y-%m-%d')

#trim volumn



```

```python
#volumn trim
target_df = target_df[target_df['volume'] != 0]
```

```python
#add pct
pct_df = target_df.copy()
convert_to_pct(pct_df, ['open', 'high', 'low', 'close'], drop_NA=True)
pct_df = pct_df.reset_index().drop('index', axis=1)

```

```python
#saving
target_df.to_csv('google_index_spec_s.csv', index=False)

#save pct
pct_df.to_csv('google_index_pct_spec_s.csv', index=False)
```
