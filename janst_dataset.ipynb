```python


#save some partitions into a big dataset


dataset = ds.dataset(r"janestreet\jane-street-real-time-market-data-forecasting\train.parquet", format="parquet", partitioning="hive")

filter_condition = (pc.field("partition_id") >= 8)

table = dataset.to_table(filter=filter_condition)
df = table.to_pandas()

# df.to_parquet("combined_train_raw.parquet", engine="pyarrow")
```

```python
#drop in place
l = ["date_id", "responder_0", "responder_1", "responder_2", "responder_3", "responder_4", "responder_5", "responder_7", "responder_8"]


df.drop(l, axis=1, inplace=True)
```

```python
#drop all nan, drop all na rows
df.dropna(axis=0, inplace=True)
```
